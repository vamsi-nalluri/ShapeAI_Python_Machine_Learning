```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.datasets import load_boston
df=load_boston()
```

```python
df.keys()
```

```python
b=pd.DataFrame(df.data, columns=df.feature_names)
b.head()
```

```python
b["MEDV"]=df.target
b.head()
```

```python
b.isnull()
```

```python
b.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X=b.drop("MEDV", axis=1)
Y=b["MEDV"]
X_train,X_test,Y_train,Y_test=train_test_split(X, Y,test_size=0.15,random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
l=LinearRegression()
l.fit(X_train,Y_train)
```

```python
y_train_predict=l.predict(X_train)
r=(np.sqrt(mean_squared_error(Y_train,y_train_predict)))
y_test_predict=l.predict(X_test)
R=(np.sqrt(mean_squared_error(Y_test,y_test_predict)))
print("The model performance for training set\nRMSE is{0}\nThe model performance for testing set\nRMSE is{1}\n".format(r,R))
```
